```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)
```

## Dataset explanation: synthetic document-like grid

The grid is not a real document. It contains simple shapes that resemble table borders, form lines, and marks.


```python
image = np.zeros((20, 20))
image[3:16, 4] = 1
image[3:16, 15] = 1
image[3, 4:16] = 1
image[15, 4:16] = 1
image[8, 4:16] = 0.8
image[3:16, 10] = 0.8
image[12:15, 2:5] = 0.7
image += np.random.normal(0, 0.03, image.shape)
image = np.clip(image, 0, 1)
```

```python
plt.figure(figsize=(5, 5))
plt.imshow(image, cmap="gray", vmin=0, vmax=1)
plt.title("Synthetic document-like grid")
plt.axis("off")
plt.show()
```

## Define filters

A filter is a small matrix of weights. It slides over local patches and gives strong responses when the local pattern matches the filter.


```python
vertical_edge_filter = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])
horizontal_edge_filter = np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]])
blur_filter = np.ones((3, 3)) / 9
sharpen_filter = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])
```

## Manual convolution/correlation function

The function extracts a local patch, multiplies it by the kernel, sums the result, and stores the response in the output feature map.


```python
def apply_filter(image, kernel, padding=1):
    padded = np.pad(image, pad_width=padding, mode="constant", constant_values=0)
    output = np.zeros_like(image, dtype=float)
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            patch = padded[i:i + kernel.shape[0], j:j + kernel.shape[1]]
            response = np.sum(patch * kernel)
            output[i, j] = response
    return output
```

```python
feature_maps = {
    "Vertical edge response": apply_filter(image, vertical_edge_filter),
    "Horizontal edge response": apply_filter(image, horizontal_edge_filter),
    "Blur response": apply_filter(image, blur_filter),
    "Sharpen response": apply_filter(image, sharpen_filter),
}
```

## Visualize feature maps

Each filter produces one feature map. Different filters respond to different local patterns.


```python
plt.figure(figsize=(10, 8))
for idx, (title, fmap) in enumerate(feature_maps.items(), start=1):
    plt.subplot(2, 2, idx)
    plt.imshow(fmap, cmap="coolwarm")
    plt.title(title)
    plt.axis("off")
    plt.colorbar(fraction=0.046, pad=0.04)
plt.tight_layout()
plt.show()
```

## One filter, one feature map

Each filter produces one feature map. A real CNN learns many filters. Multiple filters produce multiple feature maps. Later CNN layers combine simple patterns into more complex visual representations.


## Inspect one local patch response

This small numerical example shows what happens at one location.


```python
row, col = 8, 10
patch = np.pad(image, 1, mode="constant")[row:row + 3, col:col + 3]
product = patch * vertical_edge_filter
response = product.sum()
print("Patch:")
print(np.round(patch, 2))
print()
print("Kernel:")
print(vertical_edge_filter)
print()
print("Element-wise product:")
print(np.round(product, 2))
print()
print("Sum response:", round(response, 3))
```

## FinTech interpretation

In financial document processing, vertical and horizontal filters can respond to table borders. Edge filters can help locate form boundaries. Later models can combine visual patterns to classify invoices, bank statements, forms, or cheques.


## What to observe

- Filters respond strongly to certain local patterns.
- Feature maps show where patterns appear.
- Multiple filters allow a CNN to detect multiple visual cues.


## Common pitfalls

- CNN does not understand documents like humans.
- CNN learns statistical visual patterns.
- Image quality, rotation, blur, and layout changes matter.
- For pure tabular data, CNN is usually not the first choice.
